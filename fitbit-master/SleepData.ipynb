```python
def makesleep(df, resultdf) :
    for i in _ids :
        df1id = df[df['id']==i]
        for day in days :
            try :
                dfday = df1id[(df1id['time'] > day) & (df1id['time'] < day + pd.Timedelta('1 day'))]
                startsleep = dfday[dfday['inbed']!=0]['time'].iloc[0]
                minutesAsleep = dfday.iloc[0]['minutesAsleep']
                hoursAsleep = round(minutesAsleep/60,1)
                string = str(minutesAsleep) + ' minutes'
                dayindex = dfday[(dfday['time'] > startsleep) & (dfday['time'] < startsleep + pd.Timedelta(string))].index
                steps = round(df.loc[dayindex,'disturb_steps'].sum()/hoursAsleep,2)
                hr = round(df.loc[dayindex,'disturb_hr'].sum()/hoursAsleep,2)
                moving = round(df.loc[dayindex,'disturb_moving'].sum()/hoursAsleep,2)
                print(dayindex.shape, pd.Timedelta(string), hr, moving)
                df.loc[dayindex,'disturb_steps_byday'] = steps
                df.loc[dayindex,'disturb_hr_byday'] = hr
                df.loc[dayindex,'disturb_moving_byday'] = moving
                resultdf.loc[(i,day),'disturb_steps_byday'] = steps
                resultdf.loc[(i,day),'disturb_hr_byday'] = hr
                resultdf.loc[(i,day),'disturb_moving_byday'] = moving
            except : pass
```

```python
def accuracy_scorer(model, X_train, X_test, y_train, y_test):
    
    y_pred = pd.Series(data=model.predict(X_train),dtype=int)
    accuracy_train = f1_score(y_true = y_train,
                        y_pred = y_pred, average=None)
    
    y_pred = pd.Series(data=model.predict(X_test),dtype=int)
    accuracy_test = f1_score(y_true = y_test,
                       y_pred = y_pred, average=None)
    
    r2_scores = (model.score(X_train, y_train), model.score(X_test, y_test))
    
    print("R2 scores for " + str(type(model)) + " train set: " + str(r2_scores[0]))
    print("R2 scores for " + str(type(model)) + " test set: " + str(r2_scores[1]))
    print("The training accuracy-score for " + str(type(model)) + " is: ")
    print(str(accuracy_train))
    print("The testing accuracy-score for " + str(type(model)) + " is: ")
    print(str(accuracy_test))
    print()
    return (accuracy_train, accuracy_test)
```

```python
for model in [logreg_class_pca, knn_class_pca, bagged_class_pca, random_forest_class_pca, 
              adaboost_class_pca] :
    accuracy_scorer(model, X_train, X_test, y_train, y_test)
```

```python
def makeMonth(timestring) :
    return timestring[0:1]

def makeDay(timestring) :
    return int(timestring[2:4].strip('/'))

makeDay('4/2/2016 7:21:00 AM')

dfraw['month'] = dfraw['Time'].map(makeMonth)

dfraw['day'] = dfraw['Time'].map(makeDay)
dfraw.head()

def make_am_pm(timestring) :
    return timestring[-2:]

def makeSec(timestring) :
    return int(timestring[-5:-3])
    
def makeMin(timestring) :
    return int(timestring[-8:-6])
    
def makeHour(timestring) :    
    return int(timestring[-11:-9])

dfraw['hour'] = dfraw['Time'].map(makeHour)

dfraw['minute'] = dfraw['Time'].map(makeMin)

dfraw['second'] = dfraw['Time'].map(makeSec)

dfraw['am'] = dfraw['Time'].map(make_am_pm)

tempindex = dfraw[dfraw['am']=='PM'].index

dfraw.loc[tempindex,'hour'] = dfraw.loc[tempindex,'hour'].map(lambda x : x + 12)

dfraw['time_elapsed'] = ((dfraw['month'].astype(int)-4)*19*24*60*60 + 
                         (dfraw['day']-12)*24*60*60 + 
                         (dfraw['hour'])*60*60 + 
                         (dfraw['minute'])*60 +
                         (dfraw['second']))

dayinsec = 24*60*60

dfraw[dfraw['hour']>12].shape
```
